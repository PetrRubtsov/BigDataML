```scala
// необходимые для подключения библиотеки
import $ivy.`com.github.tototoshi::scala-csv:1.3.6`
import com.github.tototoshi.csv._
import java.io._
import $ivy.`org.scalanlp::breeze:1.0`
import breeze.linalg._
import breeze.numerics._
import breeze.stats.regression._
import scala.util.{Try, Success, Failure}
import com.github.fommil.netlib._
```

```scala
// предобработку признаков для линейной регрессии
val reader = CSVReader.open(new File("/home/prubtsov/ML_BigData/homework1/Dataset.csv"))
val array_list = reader.all().map(x=> x.toArray)
val dataset = DenseMatrix(array_list.map(_.map(x => Try(x.toDouble).getOrElse(1.0))):_*)
val target = dataset(::, -1)
val v = DenseVector.ones[Double](dataset(::,0).length)
val indep = DenseMatrix.horzcat( v.asDenseMatrix.t, dataset(::, 0 to 26))
```

```scala
// Построение модели простой линейной регрессии, используя стандартные функции Breeze
val regression_breeze = leastSquares(indep, target).coefficients
val resault_regression_breeze = indep * regression_breeze 
val rmse_regression_breeze = sqrt(sum((resault_regression_breeze - target)*:*(resault_regression_breeze - target))/target.length)
println("средне-квадратичное простой регрессии с помощью breeze:" + rmse_regression_breeze)
```

```scala
// Реализация линейной регрессии руками
val regression_handmade =  inv(indep.t * indep) * indep.t * target
val resault_regression_handmade = indep * regression_handmade 
val rmse_regression_handmade = sqrt(sum((resault_regression_handmade - target)*:*(resault_regression_handmade - target))/target.length)
println("средне-квадратичное простой регрессии руками:" + rmse_regression_handmade)
```
